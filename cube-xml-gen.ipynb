```python
import numpy as np
```

```python
CUBE_CONFIG_TMPLT_STR = """
				<body name="cube_{idx}" pos="{x:0.3f} {y:0.3f} 0.8">
						<inertial pos="0 0 0" mass="0.088" diaginertia="0.0001 0.0001 0.0001" />
						<geom type="box" size="0.0275 0.0275 0.0275" pos="0.0 0.0 0.0" euler=".6 .8 1" group="1" rgba="{r} {g} {b} 1" mass=".100"/>
						<site name="cube_{idx}" size="0.01" pos="0 0 0"/>
						<freejoint/>
				</body>
"""

N_CUBES = 30

X_LIMITS = (-2.0, 1.5)
Y_LIMITS = (0.425, 0.625)

COLOR_MAP = {
  "red":    (1.0, 0.0, 0.0),
  "green":  (0.0, 1.0, 0.0),
  "blue":   (0.0, 0.0, 1.0),
  "yellow": (1.0, 1.0, 0.0),
}

CUBE_CONFIG_XML = ""
for idx in range(N_CUBES):
  x = np.random.uniform(*X_LIMITS)
  y = np.random.uniform(*Y_LIMITS)
  color = list(COLOR_MAP.values())[np.random.choice(len(list(COLOR_MAP.keys())))]

  CUBE_CONFIG_XML += \
    CUBE_CONFIG_TMPLT_STR.format(
      idx=idx, x=x, y=y,
      **{k: v for k, v in zip(["r", "g", "b"], list(color))
      }
    )

print(CUBE_CONFIG_XML)
```
